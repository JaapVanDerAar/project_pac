 <div class="alert alert-success">
Here you see the average difference in correlations between the datasets (e.g. if the correlation between Amp & CF is 0.4 in humans, 0.6 in monkeys, and 0.8 in rats, the average difference is 0.2). Some are quiet high, mainly on features in which that differ per dataset. Funny to see that some values are really low; e.g. the relationship between RD symmetry and another feature does not change much between datasets.
